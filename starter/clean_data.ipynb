```python
import pandas as pd
```

# Read the data 

```python
df = pd.read_csv('data/census.csv')
df.sample(10)
```

```python
df.columns
```

Clean the data by 
- removing spaces in column names
- replace ? sign with None
- drop null rows


```python
# removing spaces in column names
df.columns = df.columns.str.strip()
```

```python
# replace ? with null values
clean_df = df.replace('\\?', None, regex=True)
# check if it is removed
clean_df.iloc[931]
```

```python
# remove all null rows
clean_df = clean_df.dropna()
```

```python
print("Number of rows before remove nulls :",df.shape[0])
print("Number of rows after remove nulls  :",clean_df.shape[0])
```

```python
# Save to new file
clean_df.to_csv('data/census_clean.csv', index=False)
```
